```python
import matplotlib.pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.legend()
plt.xlabel("Months")
plt.ylabel("Fare per month ($)")
plt.title("Avg fare per month")
plt.xlim(0,45)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis, color="red")
```

```python
#this create a scatter with the y_axis getting bigger as moving up the y_axis
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="olive", s = [i * 5 for i in y_axis], label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
plt.gca().invert_yaxis()
```
