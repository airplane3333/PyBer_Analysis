```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#list of items to plot in graph
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ["Jan","Feb", "Mar", "Apr","May","June","July","Aug","Sep","Oct","Nov","Dec"]
```

```python
#order of list maters first will be x axis
plt.plot(months , savings)
plt.xlabel("Month")
plt.ylabel ("Savings in ($)")
plt.title ("Savings Account Results")
#shows the plot chart
plt.show()
```
