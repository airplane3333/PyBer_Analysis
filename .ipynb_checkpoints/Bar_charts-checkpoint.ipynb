```python
import pandas as pd
import matplotlib.pyplot as plt 

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create a bar chart
plt.bar(x_axis , y_axis)
```

```python
#updating the attributes of the bar chart
plt.bar(x_axis, y_axis, color="purple", label="Boston")
#create labels for x / y
plt.xlabel("data")
plt.ylabel("Fare ($)")
plt.title("Per Fare by Month")
plt.legend()
```

```python
#create a horizontal bar char
plt.barh(x_axis, y_axis, color="magenta", label ="Boston")
plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.gca().invert_yaxis()
```

```python
#create an o
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black")
```

```python
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_title("Fares per month for Chicago")
#plot a gird 
ax.grid()
# Add the legend.
ax.legend()
plt.gca().invert_yaxis()
```

```python
import statistics
stdev = statistics.stdev(y_axis)
stdev
```

```python

plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots(figsize=(12,10))

ax.bar(x_axis, y_axis, yerr=stdev, capsize=3 )

plt.show()
```

```python
import numpy as np
plt.subplots(figsize=(12,10))
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
plt.grid()
```
