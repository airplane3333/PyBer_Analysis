```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#list of items to plot in graph
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#order of list maters first will be x axis
# this is for something magical but don't know why
%matplotlib inline
#import the list 
plt.plot(x_axis , y_axis)
#titles for the plot
plt.xlabel("Month")
plt.ylabel ("Total Fair in ($)")
plt.title ("Fair by month")
#shows the plot chart
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis , y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis , y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
plt.xlabel("Month")
plt.ylabel ("Total Fair in ($)")
plt.show()
plt.savefig("lineChart.png")
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="", color="red", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
#plot a gird 
plt.grid()
# Add the legend.
plt.legend()
```

```python
#seems this has to be there to make sure something is working? 
fig, ax = plt.subplots(figsize=(20,10))
# Create the plot and add a label for the legend.
ax.plot(x_axis, y_axis, marker="", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
#plot a gird 
ax.grid()
# Add the legend.
ax.legend()
```
